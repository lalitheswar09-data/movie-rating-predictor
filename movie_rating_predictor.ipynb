```python
geners=df['genres']
geners
```

```python
def format_genres(x):
    for i in x: 
            return "".join(list(i["name"]))
```

```python
g=format_genres(geners)
g
```

```python
# here the python when it read the csv files and if they contain the complex data like the list of dict
# the python read it like a string so we should use the ast 
import ast
def fix_g(x):
    try:
        x_list=ast.literal_eval(x)
        return "".join(g["name"] for g in x_list)
    except ValueError as err:
        return ""
```

```python
df['fixed_g']=df['genres'].apply(fix_g)
df['fixed_g'].head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer as cv
cv_model=cv()
vc_g=cv_model.fit_transform(df['fixed_g'])
```

```python
vc_g
```

```python
import ast
def fix_keywords(x):
    try:
        x_list=ast.literal_eval(x)
        return "".join(key["name"] for key in x_list)
    except Exception as err:
        print(err)
        return ""
```

```python
df['fixed_key']=df['keywords'].apply(fix_keywords)
df['fixed_key'].head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer as cv
cv_model=cv()
vc_key=cv_model.fit_transform(df['fixed_key'])
```

```python
# after the vecotrization the names 
# keyword vc_key
# genrrs vc_g
# title vc_title
# overivew vc title

# now i have to train the model from these vectorized featiure the ml model to get predict the rating for 10 
# the features are going to be vc and predict the rating predictor 
```

```python
from sklearn.feature_extraction.text import CountVectorizer as cv
df['overview'] = df['overview'].fillna('').astype(str)

# again the pandas read the overview as something else and 
#this needs to changed to string
cv_model =cv(stop_words='english')
vc_overview = cv_model.fit_transform(df['overview'])
vc_overview
```

```python
from scipy.sparse import hstack
# Combine all 4 vectorized feature matrices horizontally
X=hstack([vc_key, vc_g, vc_title, vc_overview])
y=df['vote_average']
```

```python
from sklearn.model_selection import train_test_split as tts
X_train,X_test,y_train,y_test=tts(X,y,test_size=0.2,random_state=18)
```

```python
from sklearn.linear_model import LinearRegression as lr
lr_model=lr()
```

```python
lr_model.fit(X_train,y_train)
```

```python
predictions=lr_model.predict(X_test)
```

```python
import numpy as np
from sklearn import metrics

# 1. Mean Absolute Error (MAE) - Average absolute distance between actual and predicted ratings
mae = metrics.mean_absolute_error(y_test, predictions)

# 2. Mean Squared Error (MSE) - Punishes larger errors more severely
mse = metrics.mean_squared_error(y_test, predictions)

# 3. Root Mean Squared Error (RMSE) - Puts the error back into the original scale (0-10 rating stars)
rmse = np.sqrt(mse)

# 4. R-squared Score (R²) - Percentage of variance explained by your text features
r2 = metrics.r2_score(y_test, predictions)

# Print the metrics nicely
print(f"Mean Absolute Error (MAE):    {mae:.4f}")
print(f"Mean Squared Error (MSE):     {mse:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
print(f"R-squared Score (R²):         {r2:.4f}")
```
